```python
#importer pandas
import pandas as pd
#importer la fonction sqrt de la bib math
from math import sqrt
# importer matplotlib
import matplotlib.pyplot as plt
#importer la fonction train_test_split du module model_selection de sklearn
from sklearn.model_selection import train_test_split
#importer la fonction LinearRegression du module linear_model de la bib sklearn
from sklearn.linear_model import LinearRegression
#importer les deux fonctions de calcul des erreurs mean_squared_error et r2_score
from sklearn.metrics import mean_squared_error, r2_score
#importer le module preprocessing de skleran
from sklearn import preprocessing
```

# **Explorez les données du fichier ozeone.cs**

```python
#lire le fichier ozone.csv
data=pd.read_csv('ozone.csv')
```

```python
data
```

# **Enlever les colonnes inutiles pour l’expérience.**

```python
data.drop(columns=["Date", "Vent","Pluie"],inplace=True)
```

# **Vérifiez s’il existe des valeurs manquantes dans les colonnes sélectionnées**

```python
data.isnull().sum()
```

# **Séparer les données prédictives et la variable à prédire**

```python
#isoler la colonne de la variable à prédire
data_y=data["MaxO3"]
data_y
```

```python
#extraire les données explicatives ou prédictives
data_x=data[['T9','T12','T15','Ne9','Ne12','Ne15','Vx9','Vx12','Vx15','MaxO3v']]
data_x
```

# **Normaliser les données prédictives pour remédier au problème des échelles des variables différentes. Nous utilisons pour cela la normalisation standard.**


```python
#Fonction permettant la normalisation des données (mise à l'échelle)
#ValeurNormaliser= valeur à normaliser-Moyenne /EcartType
def normalization(dataToNormalize):
    columns=dataToNormalize.columns
    for col in columns:
        #Extraction des valeurs de la colonne et conversion en float
        x=dataToNormalize[[col]].values.astype(float)
        #Initialisation de l'outil de normalisation
        standard_normalization = preprocessing.StandardScaler()
        #Normalisation des données
        res=standard_normalization.fit_transform(x)
        dataToNormalize[[col]]=res
```

```python
#appel à la fonction
normalization(data_x)
```

```python
data_x
```

# **Créer les ensembles des données d’entraînement et de test**

```python
#récupérer les données d'entraînement et de test
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# **Instancier l’algorithme de régression linéaire**

```python
#Créer un objet de type LinearRegression
regression_lineaire = LinearRegression()
```

# **Lancer sa fonction d’apprentissage**

```python
#Entraîner le modèle en appelant sa méthode fit en lui envoyant les données d'entraînement
regression_lineaire.fit(x_train,y_train)
```

# **Lancez les prédictions sur les données d’entraînement.**


```python
#Récupérer les prédictions du modèle de la variables à prédire en lui envoyant les variables prédictives des données d'entraînement
train_predictions = regression_lineaire.predict(x_train)
```

# **Calculez les métriques RMSE et R2 pour évaluer les prédictions sur les données d’entraînement**

```python
mean_squared_error(y_train,train_predictions)
```

```python
#Calcul du RMSE sur les données d'entraînement avec 2 chiffres après la virgule
print( f"RMSE = { round( sqrt( mean_squared_error(y_train,train_predictions) ),2)}")
#Calcul du R2_score sur les données d'entraînement
print( f"R2_score = {round(r2_score(y_train, train_predictions),2)}")
```

# **Lancez les prédictions sur les données de test**

```python
#Faire la même chose sur les données de test
test_predictions=regression_lineaire.predict(x_test)
```

# **Évaluez ces prédictions en calculant RMSE et R**

```python
print( f"RMSE = { round( sqrt( mean_squared_error(y_test,test_predictions) ),2)}")
print( f"R2_score = {round(r2_score(y_test, test_predictions),2)}")
```

```python
for i in range (100):
    x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
    regression_lineaire.fit(x_train,y_train)
    test_predictions=regression_lineaire.predict(x_test)
    print( f"Run {i} : RMSE = { round( sqrt( mean_squared_error(y_test,test_predictions) ),2)}")
    print( f"Run {i} : R2_score = {round(r2_score(y_test, test_predictions),2)}")
```

```python
from sklearn.model_selection import KFold
Kf = KFold(n_splits=4, shuffle = True)
for train_index, test_index in Kf.split(data_x):
    # print("Les indices de train index", train_index)
    # print("Les indices de test index", test_index)
    X_train, X_test = data_x.iloc[train_index],data_x.iloc[test_index]
    Y_train, Y_test = data_y.iloc[train_index],data_y.iloc[test_index]
    regression_lineaire = LinearRegression()
    regression_lineaire.fit(X_train,Y_train)
    test_predictions = regression_lineaire.predict(X_test)
    print( f"RMSE = { round( sqrt( mean_squared_error(Y_test,test_predictions) ),2)}")
    print( f"R2_score = {round(r2_score(Y_test, test_predictions),2)}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures (degree = 2)
poly_x_train = poly.fit_transform(X_train)
```

```python
print( f"RMSE = { round( sqrt( mean_squared_error(y_test,test_predictions_test) ),2)}")
print( f"R2_score = {round(r2_score(y_test, test_predictions_test),2)}")
```

```python
d=[]
r2_score_train=[]
r2_score_test=[]
for i in range (4):
    poly = PolynomialFeatures (degree = i)
    poly_x_train = poly.fit_transform(X_train)
    model = LinearRegression ()
    model.fit(poly_x_train,Y_train)
    test_predictions=model.predict(poly_x_train)
    print ("degree = ",i)
    print( f"RMSE train = { round( sqrt( mean_squared_error(Y_train,test_predictions) ),2)}")
    print( f"R2_score train = {round(r2_score(Y_train, test_predictions),2)}")

    poly_x_test = poly.fit_transform(x_test)
    test_predictions_test=model.predict(poly_x_test)
    print( f"RMSE test = { round( sqrt( mean_squared_error(y_test,test_predictions_test) ),2)}")
    print( f"R2_score test = {round(r2_score(y_test, test_predictions_test),2)}")
    d.append(i)
    r2_score_train.append(round(r2_score(Y_train, test_predictions),2))
    r2_score_test.append(round(r2_score(y_test, test_predictions_test),2))
```

```python
r2_score_train

```

```python
import matplotlib.pyplot as plt


plt.plot(d,r2_score_train,label="r2_train")
plt.plot(d,r2_score_test,label="r2_test")

plt.xlabel("degre")
plt.ylabel("R2 score")

plt.legend()
plt.show()
```
