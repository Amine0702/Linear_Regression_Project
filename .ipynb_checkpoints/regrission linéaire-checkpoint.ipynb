```python
import pandas as pd
```

```python
df=pd.read_csv("ozone.csv")
df
```

```python
df=df.drop(['Vent','Pluie'] , axis=1)
```

```python
df
```

```python
vm=df.isnull().sum()
print(vm)
```

```python
x = df[['T9', 'T12', 'T15', 'Ne9', 'Ne12', 'Ne15', 'Vx9', 'Vx12', 'Vx15', 'MaxO3v']]
y = df['MaxO3']
```

```python
from sklearn import preprocessing

def normalisation(data):
    columns = data.columns  # Utiliser 'data' au lieu de 'x'
    for col in columns:
        values = data[[col]].values.astype(float)
        sn = preprocessing.StandardScaler()
        result = sn.fit_transform(values)  # Correction ici
        data.loc[:, col] = result  # Utiliser .loc pour éviter l'avertissement
    return data
```

```python
normalisation(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
print("Coefficients du modèle : ", model.coef_)
print("Intercept du modèle : ", model.intercept_)
```

```python
import numpy as np
y_pred = model.predict(X_test)

print("Régression simple: RMSE = ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("Score R2=",r2_score(y_test,y_pred))
```

```python
y_pred_test = model.predict(X_test)

print("Prédictions sur les données de test :")
print(y_pred)
```

```python
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_test = r2_score(y_test, y_pred_test)

print(f'RMSE sur les données de test : {rmse_test}')
print(f'R² sur les données de test : {r2_test}')
```
